# Birleştirme(Join) İşlemleri

```python
import pandas as pd
import numpy as np
```

```python
m = np.random.randint(1,30,size=(5,3))
df1 = pd.DataFrame(m, columns= ["var1","var2","var3"])
df1
```

```python
#df1'e eklemek üzere başka bir dataframe oluşturalım.
df2 = df1*2
df2
```

```python
#df1 ve df2 yi birleştirmek için pd.concat() metodunu kullanacağız.

pd.concat([df1, df2])
```

```python
?pd.concat
```

```python
#Görüldüğü üzere indeksleme de birleştirilen kısım kendi indeksini kullandı bunu önlemek için "ignore_index" kullanırız.

pd.concat([df1, df2], ignore_index= True)
```

```python
#Şimdi de df2'nin sütun isimlerini değiştirelim ve öyle birleştirelim.

df2.columns = ["var1","var2","keys"]
```

```python
pd.concat([df1,df2], ignore_index= True)
```

```python
#Birleştirme'de farklı sütunlar olduğu için bazı yerlerde "NaN" yazıldı. 
#Bunun önüne geçmek için ortak sütunlar birleştirilir.
# " join="inner" " argümanı kullanılır.
pd.concat([df1, df2] , ignore_index= True, join="inner")
```
