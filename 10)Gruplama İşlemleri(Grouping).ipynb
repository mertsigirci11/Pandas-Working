# Gruplama İşlemleri(Grouping)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({"Gruplar":["A","B","C","A","C","B"],
                   "Veri": [11,17,28,54,31,38]}, columns=["Gruplar","Veri"])
df
```

Görüldüğü üzere gruplar da her bir grup için 2 tane değer var. Bunları birleştirmek için **groupby(grup_adı)** metodu kullanılır.

```python
df.groupby("Gruplar")
#Bu işlem sayesinde aynı değere sahip elemanlar birleştirildi.
```

```python
df.groupby("Gruplar").sum()
#Grup değişkenlerinin değerlerinin toplamlarını alır.
```

```python
df.groupby("Gruplar").std()
#Grup değişkenlerinin değerlerinin standart sapmalarını alır.
```
